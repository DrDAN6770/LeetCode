# 100. Same Tree
---
https://leetcode.com/problems/same-tree/description/?envType=list&envId=rwo58iji

```python
'''
 1
2 3
 1
2 3
'''
# BFS
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def isSameTree(self, p: TreeNode, q: TreeNode) -> bool:
        queue = [(p, q)]
        while queue:
            p, q = queue.pop()
            if not p and not q:
                continue
            elif (not p or not q) or (p.val != q.val):
                return False
            queue.append((p.left, q.left))
            queue.append((p.right, q.right))
        return True
ans = Solution()
p = TreeNode(1, TreeNode(None), TreeNode(3))
q = TreeNode(1, TreeNode(2), TreeNode(3))
ans.isSameTree(p, q)
```

```python
# DFS
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def isSameTree(self, p: TreeNode, q: TreeNode) -> bool:
        if not q and not p:
            return True
        elif (not p or not q) or (p.val != q.val):
            return False
        else:
            return self.isSameTree(p.left, q.left) and self.isSameTree(p.right, q.right)
ans = Solution()
p = TreeNode(1)
q = TreeNode(None)
ans.isSameTree(p, q)
```

```python
# str
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def isSameTree(self, p: TreeNode, q: TreeNode) -> bool:
        return str(p) == str(q)
ans = Solution()
p = TreeNode(1, TreeNode(None), TreeNode(3))
q = TreeNode(1, TreeNode(2), TreeNode(3))
ans.isSameTree(p, q)
```
