# 12. Integer to Roman
---
https://leetcode.com/problems/integer-to-roman/description/

```python
# O(n2) O(n)
# abaab abaab
# 01234 56789
class Solution:
    def repeatedSubstringPattern(self, s: str) -> bool:
        n = len(s)
        if n == 1:
            return False

        for i in range(1, n//2 + 1): # n//2 最少重複2次 最多重複n次
            if n % i:
                sb = s[:i]
                times = n // i
                if sb * times == s: # 驗證重複幾次
                    return True
        return False

ans = Solution()
ans.repeatedSubstringPattern("abaababaab")
```

# String Concatenation 進階用法





```python
# O(n) O(n)
class Solution:
    def repeatedSubstringPattern(self, s: str) -> bool:
        return s in s[1:] + s[:-1]

ans = Solution()
ans.repeatedSubstringPattern("abaababaab")
```
