# 518. Coin Change II
---
https://leetcode.com/problems/coin-change-ii/

```python
# O(a*n) O(a*n)
class Solution:
    def change(self, amount: int, coins: list[int]) -> int:
        if amount == 0 : return 1
        n = len(coins)
        dp = [[0] * (amount + 1) for _ in range(n + 1)]
        for i in range(n + 1):
            dp[i][0] = 1
        for c in range(1, n + 1):
            for a in range(1, amount + 1):
                if a - coins[c - 1] >= 0:
                    dp[c][a] = dp[c - 1][a] + dp[c][a - coins[c-1]]
                else:
                    dp[c][a] = dp[c - 1][a]
        return dp[-1][-1]

ans = Solution()
ans.change(5, [1,2,5])
```

```python
# O(a*n) O(a)
class Solution:
    def change(self, amount: int, coins: list[int]) -> int:
        dp = [0] * (amount + 1)
        dp[0] = 1
        for c in coins:
            for a in range(1, amount + 1):
                if a - c >= 0:
                    dp[a] += dp[a - c]
        return dp[-1]
ans = Solution()
ans.change(5, [1,2,5])
```
