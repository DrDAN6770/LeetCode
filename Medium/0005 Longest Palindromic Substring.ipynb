# 5. Longest Palindromic Substring
---
https://leetcode.com/problems/longest-palindromic-substring/?envType=list&envId=rwo58iji

```python
# 從每次index當中心往左右尋找(分中心單獨一個、無中心但左右都相同)
'''
babad(single)
1. b當中心往外找 >> b
2. a >> bab
3. b >> aba
4. a >> a
5. d >> d
--------------
cbbd(double)
1. c >> 無
2. b >> bb
3. b >> 無
4. d >> 無
'''
class Solution:
    def longestPalindrome(self, s: str) -> str:
        def check_get(s, l, r):
            while l >= 0 and r < len(s) and s[l] == s[r]:
                l -= 1
                r += 1
            return s[l+1:r]

        # 無字、同字直接返回原string
        if len(set(s)) <= 1: return s

        res = ''
        for i in range(len(s)):
            res_single = check_get(s, i, i)
            res_double = check_get(s, i, i + 1)
            res = max([res, res_single, res_double], key=len)
        return res

ans = Solution()
ans.longestPalindrome('cbbd')
```

```python
class Solution:
    def longestPalindrome(self, s: str) -> str:
        def check_get(s, l, r):
            while l >= 0 and r < len(s) and s[l] == s[r]:
                l -= 1
                r += 1
            return s[l+1:r]

        # 無字、同字直接返回原string
        if len(set(s)) <= 1: return s

        res = ''
        for i in range(len(s)):
            res_single = check_get(s, i, i)
            res_double = check_get(s, i, i + 1)
            if len(res_single) > len(res_double) and len(res_single) > len(res):
                res = res_single
            elif len(res_double) > len(res_single) and len(res_double) > len(res):
                res = res_double
        return res

ans = Solution()
ans.longestPalindrome('')
```

```python
# DP > O(n2)
class Solution:
    def longestPalindrome(self, s: str) -> str:
        n = len(s)
        dp = [[False] * n  for _ in range(n)]
        for i in range(n):
            dp[i][i] = True
        res = s[i] if n else ''

        for start_index in range(n-1, -1, -1):
            for end_index in range(start_index + 1, n):
                if s[start_index] == s[end_index]:
                    if dp[start_index + 1][end_index - 1] or end_index == start_index + 1:
                        dp[start_index][end_index] = True
                        res = max(res, s[start_index:end_index + 1], key = len)
        return res

ans = Solution()
ans.longestPalindrome(' ')
```

```python
# 暴力解
class Solution:
    def longestPalindrome(self, s: str) -> str:
        def check(s):
            return s==s[::-1]
        
        for length in range(len(s), 0, -1):
            for start_index in range(0, len(s) + 1 - length):
                if check(s[start_index:start_index + length]):
                    return s[start_index:start_index + length]


ans = Solution()
ans.longestPalindrome('a')
```
