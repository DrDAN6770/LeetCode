# 96. Unique Binary Search Trees
---
https://leetcode.com/problems/unique-binary-search-trees/

```python
# DP
class Solution:
    def numTrees(self, n: int) -> int:
        dp = [1, 1, 2, 5] + [0] * (n-3)
        for node in range(4, n + 1):
            for root in range(node):
                dp[node] += dp[root] * dp[node - root - 1]
        return dp[n]
ans = Solution()
ans.numTrees(4)
```
