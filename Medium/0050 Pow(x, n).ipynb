# 50. Pow(x, n)
---
https://leetcode.com/problems/powx-n/description/

```python
# O(logn)
class Solution:
    def myPow(self, x: float, n: int) -> float:
        return pow(x,n)
ans = Solution()
ans.myPow(2,4)
```

```python
# O(x * n)
class Solution:
    def myPow(self, x: float, n: int) -> float:
        return x**n
ans = Solution()
ans.myPow(2,4)
```

```python
# O(logn)
'''
2**5 = 2*2*2*2*2 = x*x*x*x*x
= 4 * 4 * 2      = x**2 * x**2 * x

2**4 = 2*2*2*2* = x*x*x*x
= 4 * 4     = x**2 * x**2

'''

class Solution:
    def myPow(self, x: float, n: int) -> float:
        def binary_method(x, n):
            if x == 0:
                return 0
            if n == 0:
                return 1
            if n < 0:
                return 1 / binary_method(x, n * -1)
            if n & 1:
                return x * binary_method(x * x, n // 2)
            else:
                return binary_method(x * x, n // 2)
        return binary_method(x, n)
ans = Solution()
ans.myPow(2,-2147483648)
```
