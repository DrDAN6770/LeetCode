# 86. Partition List
---
https://leetcode.com/problems/partition-list/description/

```python
# O(n) O(1)
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def partition(self, head: ListNode, x: int) -> ListNode:
        small_link, big_link = ListNode(), ListNode()
        small, big = small_link, big_link # dummy

        while head:
            if head.val < x:
                small.next = head
                small = small.next # update
            else:
                big.next = head
                big = big.next
            
            head = head.next
        
        big.next = None # break link with head
        small.next = big_link.next # 相接
        
        return small_link.next

def createnode(nums):
    my_link = ListNode()
    curr = my_link
    for n in nums:
        curr.next = ListNode(n)
        curr = curr.next
    return my_link.next

mylink = createnode([1,4,3,2,5,2])
ans = Solution()
r = ans.partition(mylink, 3)
```

```python
while r:
    print(r.val, end = ' ')
    r = r.next
```
