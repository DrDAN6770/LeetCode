# 11. Container With Most Water


---

https://leetcode.com/problems/container-with-most-water/description/?envType=study-plan-v2&envId=leetcode-75

image.png

```python
class Solution:
    def maxArea(self, height: list[int]) -> int:
        l = 0
        r = len(height)-1
        currmax = 0
        while l < r:
            min_h = min(height[l], height[r])
            currmax = max(currmax, (r-l) * min_h)
            if height[l] < height[r]:
                l += 1
                while min_h > height[l]: # 移動後的新高度比原本最小高度高才有需要去重算最大面積，如果新高度比舊高度小，且寬又縮小，新面積一定更小
                    l += 1
            else:
                r -= 1
                while min_h > height[r]: # 同理
                    r -= 1
        return currmax
height = [1,8,6,2,5,4,8,3,7]
res = Solution()
res.maxArea(height)
```
