```python
# O(n) O(n)
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
def create(lst):
    dummy = ListNode()
    cur = dummy
    for i in lst:
        cur.next = ListNode(i)
        cur = cur.next
    return dummy.next

class Solution:
    def reorderList(self, head: ListNode) -> None:
        """
        Do not return anything, modify head in-place instead.
        """
        # save a list
        lst = []
        dummy = head
        while dummy:
            lst.append(dummy)
            dummy = dummy.next

        # 左右指針互指
        # 1234
        # 1 > 4 > 2 > 3
        l, r = 0, len(lst) - 1
        cur = head

        while l < r:
            lst[l].next = lst[r]
            l += 1

            if l == r:
                cur = lst[l]
                break

            lst[r].next = lst[l]
            r -= 1

            cur = lst[l]
        
        # tail指向None
        cur.next = None
```

```python
# O(n) O(1)
'''
1 2 3 4 5

1st:
1 2

2nd:
4 3

1 > 4 > 2 > 3 > None
'''
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def reorderList(self, head: ListNode) -> None:
        """
        Do not return anything, modify head in-place instead.
        """
        # 找中間斷點
        slow = fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

        # 反轉後半部
        second = slow
        prev = None
        while second:
            next_node = second.next
            second.next = prev
            prev = second
            second = next_node
        
        # 處理轉換
        first = head
        second = prev

        while second.next:
            temp1 = first.next
            temp2 = second.next

            first.next = second
            second.next = temp1

            first = temp1
            second = temp2
```
