# 2141. Maximum Running Time of N Computers
---
https://leetcode.com/problems/maximum-running-time-of-n-computers/description/

```python
# 有一堆電池，同時跑n台電腦，同時可以跑最長多久?
# 理論上最大可跑時間是 總電量/電腦數
# 最大電量的電池給固定一台電腦 >> 總電量扣最大電池, 電腦數-1
class Solution:
    def maxRunTime(self, n: int, batteries: list[int]) -> int:
        if len(batteries) == n:
            return min(batteries)
        if n == 1:
            return sum(batteries)
        
        batteries.sort()
        total_power = sum(batteries)
        while batteries[-1] > total_power // n:
            total_power -= batteries.pop()
            n -= 1
        return total_power // n
ans = Solution()
ans.maxRunTime(3,[10,10,3,5])
```

```python
class Solution:
    def maxRunTime(self, n: int, batteries: list[int]) -> int:
        if len(batteries) == n:
            return min(batteries)
        if n == 1:
            return sum(batteries)
        
        # 所有電池中 小於中間值的 平均電量 和 與中間值 的關係
        l = min(batteries)
        r = sum(batteries) // n
        while l < r:
            m = r - (r-l) // 2
            if sum(min(power, m) for power in batteries) // n >= m:
                l = m
            else:
                r = m - 1
        return l
ans = Solution()
ans.maxRunTime(3,[10,10,3,5])
```
