# 664. Strange Printer
---
https://leetcode.com/problems/strange-printer/

```python
# Bottom-Up DP
class Solution:
    def strangePrinter(self, s: str) -> int:
        if len(set(s)) == 1:
            return 1
        n = len(s)
        dp = [[0] * n for _ in range(n)]

        for i in range(n-1, -1, -1):
            dp[i][i] = 1
            for j in range(i + 1, n):
                if s[i] == s[j]:
                    dp[i][j] = dp[i][j-1]
                else:
                    dp[i][j] = float('inf')
                    for k in range(i, j):
                        dp[i][j] = min(dp[i][j], dp[i][k] + dp[k + 1][j])
        return dp[0][-1]      
        
ans = Solution()
ans.strangePrinter("aaabc")
```

```python
class Solution:
    def strangePrinter(self, s: str) -> int:
        if len(set(s)) == 1:
            return 1
        n = len(s)
        dp = [[0] * n for _ in range(n)]

        for i in range(n):
            dp[i][i] = 1

        for length in range(2, n + 1):
            for left in range(n - length + 1):
                right = left + length - 1
                dp[left][right] = length # 

                for k in range(left, right):
                    turns = dp[left][k] + dp[k+1][right]
                    if s[k] == s[right]:
                        turns -= 1
                    dp[left][right] = min(dp[left][right], turns)
        return dp[0][-1]

solution = Solution()
print(solution.strangePrinter("abab"))
```
